```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv(
    "data.csv",
    encoding="latin-1",
    engine="python",
    on_bad_lines="skip"
)
```

```python
book.info()
```

```python
print(book.columns)
```

```python
df = book[[ 'title', 'authors', 'categories','published_year']]
```

```python
df.info()
```

```python
df
```

```python
df = df.iloc[:2000]
```

```python
print(df.shape)
```

```python
# Change the datatype of published_year 
df["published_year"] = pd.to_numeric(df["published_year"], errors = 'coerce')   
df.dropna(subset=["published_year"], inplace = True )    # Drop the null data  
df["published_year"] = df["published_year"].astype(int)          #converted the data into integer formate 
```

```python
 df.fillna(" ", inplace = True)
```

```python
#Create the function that is used to make the text in lowercase , remove stop words , remove punctiontion
df.columns
```

```python
df.duplicated().value_counts()   #Check duplicate values 
```

```python
# to remove space in between words 
#df['title'] = df['title'].str.replace(' ' ,'_')
df['authors'] = df['authors'].str.replace(' ' ,'_')
df['categories'] = df['categories'].str.replace(' ' ,'_')
```

```python
# inserted the column of ID so that proper indexing the data will get 
df.insert(0, "ID", range(1, len(df) + 1))
```

```python
df
```

```python
df['tags'] = (
    df['authors'] + ',' +
    df['categories'] + ',' +
    df['published_year'].astype(str)
)
```

```python
df['tags'] 
```

```python
df.columns
```

```python
df =df[['ID', 'title',  'tags']]
```

```python
df.head()
```

```python
df['tags']= df['tags'].str.lower()    # convert everything in to lowercase 
```

```python
df['tags'] = df['tags'].str.replace('.' ,'')
```

```python
df.to_csv("cleaned_data.csv")
```
